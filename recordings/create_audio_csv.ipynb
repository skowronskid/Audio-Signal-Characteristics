```python
import sys
import glob
import pandas as pd
import warnings
from tqdm.auto import tqdm 
sys.path.append('..')
from functions.time_domain import *
from functions.frequency_domain import *
from functions.mfcc import *
```

```python
def get_all_audio_paths():
    search_pattern = f"comparison/**/Znormalizowane/*.wav"
    file_paths = glob.glob(search_pattern, recursive=True)
    return file_paths


def get_data(file_path):
    split_path = file_path.split("/")
    word_label = split_path[-1].split("_")[0]
    person = split_path[-3]
    return word_label, person
```

```python
tensor_df = pd.DataFrame(columns=['word_label', 'person', 'mfcc', 'energy'])
file_paths = get_all_audio_paths()
warnings.filterwarnings("ignore")
for file_path in tqdm(file_paths):
    try:
        word_label, person = get_data(file_path)
        mfcc, energy = mfcc_pipeline(file_path)
        
        tensor_df = tensor_df.append({'word_label': word_label, 'person': person, 'mfcc': mfcc, 'energy': energy}, ignore_index=True)
    except Exception as e:
        print(f"Something went wrong with {file_path}")
        continue
warnings.filterwarnings("default")
```

```python
tensor_df.to_pickle("audio_df.pkl")
```
